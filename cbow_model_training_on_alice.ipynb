```python
import spacy
nlp = spacy.load('en_core_web_sm')
```

```python
with open('ALICE.txt', 'r') as file:
    text = file.read()
```

```python
doc = nlp(text.lower().replace("\n", ""))
```

```python
data = []
for token in doc:
    data.append(token.text)
```

```python
data = [elem for elem in data if elem.isalnum()]
```

```python
from gensim.models import Word2Vec
import logging
import nltk
from nltk.tokenize import sent_tokenize, word_tokenize

# Set up logging
logging.basicConfig(format='%(asctime)s : %(levelname)s : %(message)s', level=logging.INFO)

# Read the corpus
with open("ALICE.txt", "r") as file:
    raw_text = file.read()

# Preprocess the text
text = raw_text.lower()

# Tokenize into sentences and words
sentences = sent_tokenize(text)
tokenized_sentences = [word_tokenize(sentence) for sentence in sentences]

# Remove alphanumerical words
tokenized_sentences_filtered = [[word for word in sentence if word.isalpha()] for sentence in tokenized_sentences]

# Instantiate CBOW model using Word2Vec
cbow_model = Word2Vec(tokenized_sentences_filtered, min_count=1, vector_size=100, window=2, sg=0, workers=4, compute_loss=True)

# Train the model
cbow_model.train(tokenized_sentences_filtered, total_examples=len(tokenized_sentences_filtered), epochs=200, compute_loss=True)
```

```python
# Get the vector representation of a word
vector = cbow_model.wv['beginning']
print(vector)

# Print out the vocabulary
vocab = cbow_model.wv.key_to_index
print(vocab)
```
