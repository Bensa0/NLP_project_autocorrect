```python
from keras.preprocessing.text import Tokenizer
from keras_preprocessing.sequence import pad_sequences

# Define the text dataset
texts = ["The cat sat on the mat", "The dog chased the cat", "The mouse ran up the clock"]

# Tokenize the text
tokenizer = Tokenizer()
tokenizer.fit_on_texts(texts)
sequences = tokenizer.texts_to_sequences(texts)

# Pad the sequences to a fixed length
max_length = max(len(seq) for seq in sequences)
padded_sequences = pad_sequences(sequences, maxlen=max_length)

# Define the vocabulary size and embedding dimension
vocab_size = len(tokenizer.word_index) + 1
embedding_dim = 50
```

```python
from keras.models import Sequential
from keras.layers import Embedding

# Define the model
model = Sequential()
model.add(Embedding(input_dim=vocab_size, output_dim=embedding_dim, input_length=max_length))

# Print the model summary
model.summary()
```

```python
# Define a new text sample
new_texts = ["The cat chased the mouse", "The dog sat on the mat"]

# Tokenize and pad the new sequences
new_sequences = tokenizer.texts_to_sequences(new_texts)
new_padded_sequences = pad_sequences(new_sequences, maxlen=max_length)

# Use the model to make predictions
predictions = model.predict(new_padded_sequences)

# Print the predictions
print(predictions)
```
