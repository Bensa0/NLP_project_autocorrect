```python
words_list = []
with open("2of12id.txt", "r") as f:
    for line in f:
        # Remove leading/trailing whitespaces
        line = line.strip()
        
        # Split the line into words
        words = line.split()
        
        # Keep only the words that are alphabetic
        alpha_words = [w.lower() for w in words if w.isalpha() and len(w) > 1]
        
        # Add the alpha words to the list
        words_list.append(alpha_words)
        
print(words_list)
```

```python
from gensim.models import FastText

def autocorrect(text, fasttext_model):
    # Tokenize the text
    tokens = text.split()

    # Initialize an empty list to store corrected tokens
    corrected_tokens = []

    # Loop through the tokens
    for token in tokens:
        # Check if the token is in the FastText vocabulary
        if token in fasttext_model.wv.key_to_index:
            corrected_tokens.append(token)
        else:
            # If token is not in the vocabulary, get the embedding from FastText
            embedding = fasttext_model.wv[token]
            # Find the most similar word to the OOV token
            similar_word = fasttext_model.wv.similar_by_vector(embedding)[0][0]
            corrected_tokens.append(similar_word)

    # Join the corrected tokens back into a sentence
    corrected_text = ' '.join(corrected_tokens)

    return corrected_text
```

```python
import random

# function to create misspelled versions of a word
def create_misspelled(word):
    # create a list to store the misspelled versions
    misspelled_list = []
    
    # add different types of errors to the list
    # missing letters
    for i in range(len(word)):
        misspelled = word[:i] + word[i+1:]
        misspelled_list.append(misspelled)
    
    # repeated letters
    for i in range(len(word)):
        misspelled = word[:i] + word[i] + word[i:]
        misspelled_list.append(misspelled)
        
    # swapped adjacent letters
    for i in range(len(word)-1):
        misspelled = word[:i] + word[i+1] + word[i] + word[i+2:]
        misspelled_list.append(misspelled)
    
    # replace letters with random letters
    for i in range(len(word)):
        misspelled = word[:i] + random.choice('abcdefghijklmnopqrstuvwxyz') + word[i+1:]
        misspelled_list.append(misspelled)
    
    return misspelled_list

# example corpus
corpus = ['apple', 'banana', 'cherry', 'dog', 'elephant', 'frog']

# create dictionary of misspelled words
misspelled_dict = {}
for word in corpus:
    misspelled_dict[word] = create_misspelled(word)
```

```python
import random

# Define the list of fruits
fruits = ['apple', 'banana', 'orange', 'kiwi', 'mango', 'papaya', 'pineapple', 'grape', 'pear', 'peach']

# Define a function to generate misspelled words
def generate_misspelled(word):
    # Split the word into characters
    chars = list(word)
    # Shuffle the characters
    random.shuffle(chars)
    # Join the shuffled characters back into a string
    misspelled = ''.join(chars)
    # Return the misspelled word
    return misspelled

# Create a list of lists to hold the words and their misspellings
words_list = [[word] + [generate_misspelled(word) for i in range(100)] for word in fruits]
words_list
```

```python
# Import FastText module from gensim package
from gensim.models import FastText

# Create pairs of correct and incorrect words
word_pairs = []
for word_group in words_list:
    for i, correct_word in enumerate(word_group):
        for j, incorrect_word in enumerate(word_group):
            if i != j:
                word_pairs.append([correct_word, incorrect_word])
word_pairs
```

```python
# Import FastText module from gensim package
from gensim.models import FastText

# Create pairs of correct and incorrect words
word_pairs = []
for word_group in words_list:
    correct_word = word_group[0]
    for incorrect_word in word_group[1:]:
        word_pairs.append([correct_word, incorrect_word])

# Train FastText model on word pairs
model = FastText(vector_size=100, window=5, min_count=0)
model.build_vocab(corpus_iterable=word_pairs)
model.train(corpus_iterable=word_pairs, total_examples=len(word_pairs), epochs=10)

# Get most similar words and similarity score to the correct word
input_word = 'plaep'  # Example input
similar_words = model.wv.most_similar(input_word, topn=5)

# Find the correct word for the input_word
correct_word = None
for word_group in words_list:
    if input_word in word_group[1:]:
        correct_word = word_group[0]
        break

if correct_word:
    similarity_score = model.wv.similarity(correct_word, input_word)
    print("Similar words:", similar_words)
    print(f"Similarity score with '{correct_word}':", similarity_score)
else:
    print("Input word not found in words_list")
```
